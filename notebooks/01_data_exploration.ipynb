```python
import os
import pandas as pd   # <-- import pandas
import kagglehub

# Download dataset
path = kagglehub.dataset_download("neuromusic/avocado-prices")
print("Path to dataset files:", path)

# Load CSV (the dataset has "avocado.csv")
df = pd.read_csv(os.path.join(path, "avocado.csv"))
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

print(f"Dataset shape: {df.shape}")
print("\nColumns:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
print("\nBasic statistics:\n", df.describe())


df['Date'] = pd.to_datetime(df['Date'])
print(f"\nDate range: {df['Date'].min()} to {df['Date'].max()}")


print("\nRegions:", df['region'].unique())
print("\nTypes:", df['type'].unique())


plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='AveragePrice', data=df)
plt.xticks(rotation=90)
plt.title('Avocado Price Distribution by Region')
plt.tight_layout()
plt.savefig("price_by_region.png")

plt.show()


national_avg = df.groupby('Date')['AveragePrice'].mean()
plt.figure(figsize=(12, 6))
national_avg.plot()
plt.title('National Average Avocado Price Over Time')
plt.ylabel('Average Price')
plt.tight_layout()
plt.savefig('price_over_time.png')
plt.show()
```
