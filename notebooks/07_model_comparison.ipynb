```python
import json
import matplotlib.pyplot as plt

# Load metrics from all models
with open('../models/baseline_metrics.json', 'r') as f:
    baseline = json.load(f)

with open('../models/prophet_metrics.json', 'r') as f:
    prophet = json.load(f)

with open('../models/xgboost_metrics.json', 'r') as f:
    xgboost = json.load(f)

# Create comparison
models = [baseline, prophet, xgboost]
model_names = [m['model'] for m in models]
mae_scores = [m['mae'] for m in models]
rmse_scores = [m['rmse'] for m in models]

# Plot comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# MAE comparison
ax1.bar(model_names, mae_scores, color=['blue', 'orange', 'green'])
ax1.set_title('MAE Comparison')
ax1.set_ylabel('Mean Absolute Error')

# RMSE comparison
ax2.bar(model_names, rmse_scores, color=['blue', 'orange', 'green'])
ax2.set_title('RMSE Comparison')
ax2.set_ylabel('Root Mean Squared Error')

plt.tight_layout()
plt.savefig('../models/model_comparison.png')
plt.show()

# Print results
print("Model Comparison:")
print("=================")
for model in models:
    print(f"{model['model'].upper():<10} MAE: {model['mae']:.4f} RMSE: {model['rmse']:.4f}")
```
